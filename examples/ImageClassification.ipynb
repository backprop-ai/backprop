# Backprop Core Example: Zero-Shot Image Classification

## A brief intro

Zero-shot classification is a relatively simple idea.
As with standard classification, a model looks at input and assigns probabilities to a set of labels.
However, with zero-shot, the model was not trained on any particular set of labels.
This makes the classifier extremely flexible for a variety of use cases.

```python
# Leave it as None to run locally
api_key = None
```

```python
import backprop

ic = backprop.ImageClassification(api_key=api_key)

# An example set of labels, to do with classifying
labels = ['healthy brain', 'brain with tumor']
```

```python
from PIL import Image

image_path = "data/brain.jpeg"
# Preview
Image.open(image_path).resize((200, 200), Image.NEAREST)
```

```python
results = ic(image_path, labels)
print(results)

label = max(results, key = results.get)
print("Image is:", label)
```
